# Verification of My Bachelor's Degree

## Business Administration and Engineering: Electrical Power Engineering (B.Sc. RWTH)

I completed my Bachelor of Science at RWTH Aachen University on **20 October 2022**. The degree comprises 180 credit points across three main areas: mathematical and natural sciences, engineering sciences, and economics and business sciences. My overall grade is **3.5 (satisfactory)**.

This notebook verifies the official documents (`urkunde.pdf` and `detail.pdf`) against the structured metadata in `key_details.csv`. It also explains my course structure, grades, and how they reflect my foundation in electrical power engineering and related fields.

```python
import pandas as pd
import os
from pypdf import PdfReader

# Define paths to the official documents and metadata
BASE_PATH = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
URKUNDE_PATH = os.path.join(BASE_PATH, 'urkunde.pdf')
DETAIL_PATH = os.path.join(BASE_PATH, 'detail.pdf')
KEY_DETAILS_CSV = os.path.join(BASE_PATH, 'data', 'key_details.csv')

def extract_text_from_pdf(pdf_path):
    """Extract full text from PDF for verification."""
    full_text = ""
    try:
        reader = PdfReader(pdf_path)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                full_text += page_text + "\n"
    except Exception as e:
        print(f"Error reading PDF: {e}")
    return full_text.lower()

# Extract text from both official documents
urkunde_text = extract_text_from_pdf(URKUNDE_PATH)
detail_text = extract_text_from_pdf(DETAIL_PATH)
full_pdf_text = urkunde_text + detail_text

print("PDF Extraction Complete")
print("=" * 50)
print(f"Total characters extracted: {len(full_pdf_text)}")
print("First 600 characters for reference (urkunde.pdf):")
print(urkunde_text[:600])
```

### Verification Process

I verify every recorded detail (names, dates, grades, thesis topic, examiner, signatures) against the full text of both official PDFs. This ensures complete alignment between the structured metadata and the original documents.

```python
def verify_detail(section, key, expected_value):
    if expected_value is None or str(expected_value).strip() == "":
        return True
    search_term = str(expected_value).strip().lower()
    if search_term in full_pdf_text:
        print(f"[VERIFIED] {section} → {key}: '{expected_value}' confirmed in official documents.")
        return True
    else:
        print(f"[ISSUE] {section} → {key}: '{expected_value}' not found.")
        return False

verification_results = []
for _, row in key_details_df.iterrows():
    section = row['section']
    key = row['detail_key']
    value = row['detail_value']
    result = verify_detail(section, key, value)
    verification_results.append(result)

total = len(key_details_df)
passed = sum(verification_results)
print("\n" + "=" * 60)
print(f"VERIFICATION SUMMARY: {passed}/{total} details confirmed")
print("=" * 60)
```
